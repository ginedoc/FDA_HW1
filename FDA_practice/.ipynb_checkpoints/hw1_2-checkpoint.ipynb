## 1. Overall and CLub

```python
import numpy as np
import scipy
import pandas as pd
```

```python
data_path = "../input/fifa.csv"
pd_data = pd.read_csv(data_path,encoding = "ISO-8859-1")
pd_data = pd_data[:10000]
pd_data.head(10)
```

## Overall 

```python
mean = pd_data.groupby(['Club']).mean()
overall = mean.sort_values(by=['Club']).Overall
potential = mean.sort_values(by=['Club']).Potential

result = pd.concat([overall, potential], axis=1)

print(result)
```

# potential

```python
result_ratio = result.Potential / result.Overall
result_ratio.plot()
```
